# Age Distribution of Customers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reading Data
data = pd.read_csv("Cust_Purch_FakeData.csv")


plt.figure(figsize=(10,5))
sns.histplot(data['age'], bins=20, kde=True, color='blue')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

# Spending Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(data['price(CAD)'], bins=30, kde=True, color='purple')
plt.title('Spending Distribution (CAD)')
plt.xlabel('Amount Spent')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

 # Most Common Customer Names

```python
top_names = data['first'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_names.index, y=top_names.values, palette='viridis', hue=top_names.index, legend=False)
plt.title('Top 10 Most Common First Names')
plt.xlabel('First Name')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

# Structural Engineers by Gender

```python
gender_counts = data[data['profession'] == 'Structural Engineer']['gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['pink', 'lightblue'], startangle=140)
plt.title('Gender Distribution of Structural Engineers')
plt.show()
```

# Spending vs. Age Scatter Plot

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=data['age'], y=data['price(CAD)'], hue=data['gender'], alpha=0.7)
plt.title('Spending Amount vs. Age')
plt.xlabel('Age')
plt.ylabel('Spending Amount (CAD)')
plt.legend(title="Gender")
plt.grid()
plt.show()
```

#  Most Active Shopping Days

```python
day_counts = data['weekday'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=day_counts.index, y=day_counts.values, palette='viridis', hue=day_counts.index, legend=False)
plt.title('Most Active Shopping Days')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

# Most Popular Email Providers

```python
email_providers = data['email'].apply(lambda x: x.split('@')[1]).value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=email_providers.index, y=email_providers.values, palette='magma', hue=email_providers.index, legend=False)
plt.title('Top 10 Email Providers Used by Customers')
plt.xlabel('Email Provider')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

# Credit Card Type Distribution

```python
data['cc_type'] = data['cc_type'].str.replace(' ', '\n')
data['cc_type'] = data['cc_type'].replace('Diners\nClub\nUnited\nStates\n&\nCanada', 'Diners\nUS\n&Canada')
data['cc_type'] = data['cc_type'].replace('Diners\nClub\nCarte\nBlanche', 'Diners\nCarte\nBlanche')
cc_types = data['cc_type'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=cc_types.index, y=cc_types.values, palette='Set2', hue=cc_types.index, legend=False)
plt.title('Distribution of Credit Card Types')
plt.xlabel('Credit Card Type')
plt.ylabel('Number of Customers')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
